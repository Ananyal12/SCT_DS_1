```python
import matplotlib.pyplot as plt

# Data
age_groups = ['0–9', '10–19', '20–29', '30–39', '40–49', '50–59', '60–69', '70–79', '80+']
population = [242, 270, 230, 195, 160, 125, 85, 55, 25]
colors = ['#FFD700'] * 2 + ['#1E90FF'] * 4 + ['#FF69B4'] * 3  # Yellow = young, Blue = working, Pink = older

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(age_groups, population, color=colors)

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, f'{height} Mn', ha='center', fontsize=9)

plt.title("India's Population by Age Group (2022)")
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
